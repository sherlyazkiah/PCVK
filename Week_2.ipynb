1. By using figsize, notice whether the image pixel size also changes?

```python
plt.figure(figsize=(5,5))   # smaller figure
plt.imshow(img)
plt.title("Small Figure Size")
plt.show()

plt.figure(figsize=(10,8))  # larger figure
plt.imshow(img)
plt.title("Large Figure Size")
plt.show()
```

2.	Show images in Red-Green and Green-Blue channels only!

```python
b, g, r = cv.split(img)

# Red-Green only (Blue = 0)
rg = cv.merge([np.zeros_like(b), g, r])

# Green-Blue only (Red = 0)
gb = cv.merge([b, g, np.zeros_like(r)])

plt.subplot(1,2,1); plt.imshow(cv.cvtColor(rg, cv.COLOR_BGR2RGB)); plt.title("Red-Green")
plt.subplot(1,2,2); plt.imshow(cv.cvtColor(gb, cv.COLOR_BGR2RGB)); plt.title("Green-Blue")
plt.show()
```

3.	Show image row 10-100, column 10-100!

```python
cropped_img = img[10:100, 10:100]
plt.imshow(cv.cvtColor(cropped_img, cv.COLOR_BGR2RGB))
plt.show()
```

4.	Show image rows 5-30, all columns, channel Green only!

```python
green_img = img[5:30, :, 1] #take green channel, rows 5-30, all columns
plt.imshow(green_img, cmap='gray')
plt.show()
```

5.	Create 5 boxes of different sizes and colors in one image. it is recommended to use random numbers!

```python
canvas = np.zeros((512,512,3), dtype=np.uint8)

for i in range(5):
    pt1 = (np.random.randint(0,400), np.random.randint(0,400))
    pt2 = (pt1[0]+np.random.randint(20,100), pt1[1]+np.random.randint(20,100))
    color = (np.random.randint(0, 256), np.random.randint(0, 256), np.random.randint(0, 256))
    cv.rectangle(canvas, pt1, pt2, color, -1)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
plt.title("Random 5 Boxes")
plt.show()
```

6.	Show the image upside down!

```python
flipped = cv.flip(canvas, 0)   # 0 = vertical flip
plt.imshow(cv.cvtColor(flipped, cv.COLOR_BGR2RGB))
plt.title("Upside Down")
plt.show()
```
